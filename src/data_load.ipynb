#### Renvoie d'un DataFrame propre

```python
def clean_dataframe(df):
    print("Nettoyage du DataFrame...")

    # enlever les NaN dans le texte
    df = df.dropna(subset=["cv_text"])

    # remplacer NaN dans les labels (option)
    df["job_domain"] = df["job_domain"].fillna("Unknown")
    df["experience_level"] = df["experience_level"].fillna("Unknown")
    df["skills"] = df["skills"].fillna("Unknown")

    # supprimer doublons
    df = df.drop_duplicates()

    print("Nettoyage terminé.")
    return df
```

```python
df = clean_dataframe(df)
```
